#### 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate synthetic data
# Feature 1: Random values
feature_1 = np.random.normal(loc=10, scale=5, size=100)

# Feature 2: Correlated with Feature 1
feature_2 = feature_1 + np.random.normal(loc=0, scale=2, size=100)

# Feature 3: Correlated with both Feature 1 and Feature 2
feature_3 = feature_1 + feature_2 + np.random.normal(loc=0, scale=1, size=100)

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Compute the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.5)

# Adding labels to the bubbles
for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=9, ha='center')

# Set plot title and labels
plt.title('Bubble Chart of Country Population vs GDP')
plt.xlabel('GDP (in billion USD)')
plt.ylabel('Population (in millions)')

# Show the plot
plt.grid()
plt.show()
```

#### 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate the sine of x values
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)

# Add a bar renderer
p.vbar(x=fruits, top=counts, width=0.5, color="blue", legend_label="Counts")

# Customize the legend
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)  # 1000 random numbers from a normal distribution

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", alpha=0.7)

# Show the plot
show(p)
```

#### 5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)â€¨
x = np.linspace(0, 1, 10)â€¨
y = np.linspace(0, 1, 10)â€¨
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
data_heatmap = np.random.rand(10, 10)  # 10x10 matrix of random values
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh figure
p = figure(title="Heatmap of Random Data", x_axis_label='X', y_axis_label='Y', width=500, height=500)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add squares to the heatmap
p.rect(x=xx.flatten(), y=yy.flatten(), width=0.1, height=0.1, source={'value': data_heatmap.flatten()}, 
        fill_color=mapper, line_color='black')

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
